```python
summary_merged_df = pd.read_csv(
    "../dataset/analytics-csv/Summary Merged v1.csv",
    converters={
        'outputs': json.loads,
        'run': json.loads
    }
)

summary_merged_df['conversation_summary'] = summary_merged_df['outputs'].apply(
    lambda x: x['conversation_summary']
)

summary_merged_df['human_values'] = summary_merged_df['outputs'].apply(
    lambda x: x.get('human_values')
)
summary_merged_df['chatbot_recommendations'] = summary_merged_df['outputs'].apply(
    lambda x: x.get('chatbot_recommendations')
)
summary_merged_df['chatbot_response_type'] = summary_merged_df['outputs'].apply(
    lambda x: x.get('chatbot_response_type')
)
summary_merged_df['bot_name'] = summary_merged_df['run'].apply(
    lambda x: x.get('extra', {}).get('metadata', {}).get('ls_example_bot_name')
)
summary_merged_df['session_id'] = summary_merged_df['run'].apply(
    lambda x: x.get('extra', {}).get('metadata', {}).get('ls_example_session_id')
)

summary_merged_df['user_id'] = summary_merged_df['run'].apply(
    lambda x: x.get('extra', {}).get('metadata', {}).get('ls_example_user_id')
)

summary_merged_df = summary_merged_df[[
    'session_id', 'user_id', 'bot_name', 'conversation_summary', 'human_values', 'chatbot_recommendations', 'chatbot_response_type'
]]

summary_merged_df.head()
```

```python
abandonment_analysis_df = pd.read_csv(
    "../dataset/analytics-csv/Abandonment Analysis O3.csv",
    converters={'outputs': json.loads,
                'run': json.loads}
)

# Extract all keys from the 'outputs' dict and add as columns
# Extract specific fields from the 'outputs' dict and add as columns
abandonment_analysis_df['abandonment_category'] = abandonment_analysis_df['outputs'].apply(lambda x: x.get('abandonment_category'))
abandonment_analysis_df['abandonment_subcategory'] = abandonment_analysis_df['outputs'].apply(lambda x: x.get('abandonment_subcategory'))
abandonment_analysis_df['abandonment_reason'] = abandonment_analysis_df['outputs'].apply(lambda x: x.get('abandonment_reason'))
abandonment_analysis_df['number_of_completed_steps'] = abandonment_analysis_df['outputs'].apply(lambda x: x.get('number_of_completed_steps'))


# Extract metadata from 'run'->'extra'->'metadata' if present
def get_metadata_field(run, field):
    return run.get('extra', {}).get('metadata', {}).get(field)

abandonment_analysis_df['bot_name'] = abandonment_analysis_df['run'].apply(lambda x: get_metadata_field(x, 'ls_example_bot_name'))
abandonment_analysis_df['session_id'] = abandonment_analysis_df['run'].apply(lambda x: get_metadata_field(x, 'ls_example_session_id'))
abandonment_analysis_df['user_id'] = abandonment_analysis_df['run'].apply(lambda x: get_metadata_field(x, 'ls_example_user_id'))

abandonment_analysis_df = abandonment_analysis_df[['session_id', 'user_id', 'bot_name', 'abandonment_category', 'abandonment_subcategory', 'abandonment_reason', 'number_of_completed_steps']]


abandonment_analysis_df.head()
```
